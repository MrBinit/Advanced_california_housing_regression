```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame named df
# Replace 'df' with your DataFrame containing numerical data

# Calculate correlation matrix
corr_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(20, 20))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python

# Calculate correlation matrix
corr_matrix = df.corr()

# Filter correlation matrix to include only positive correlations with 'SalesPrice' above a threshold
threshold = 0.5  # Adjust threshold as 0.5
positive_corr = corr_matrix['SalePrice'][corr_matrix['SalePrice'] > threshold]

# Select features with positive correlation and strong relationship with 'SalesPrice'
selected_features = positive_corr.index.tolist()

# Display selected features
print("Features with positive correlation and strong relationship with 'SalesPrice':")
print(selected_features)

```

```python
# Visualize selected features with a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df[selected_features].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Selected Features with SalePrice')
plt.show()
```

```python


# Selecting the numerical columns and 'SalePrice'
selected_data = df[selected_features]

# Determine the number of numerical variables
num_numerical_vars = len(selected_data.columns) - 1  # Excluding 'SalePrice'

# Determine the number of subplot rows and columns
num_subplot_rows = (num_numerical_vars - 1) // 3 + 1
num_subplot_cols = min(num_numerical_vars, 3)

# Plotting scatter plots for each numerical variable against 'SalePrice'
plt.figure(figsize=(16, 5 * num_subplot_rows))
for i, column in enumerate(selected_data.columns[:-1]):
    plt.subplot(num_subplot_rows, num_subplot_cols, i + 1)  # Creating subplots
    sns.scatterplot(x=column, y='SalePrice', data=selected_data)
    plt.title(f'Scatter plot of {column} vs SalePrice')

plt.tight_layout()
plt.show()
```
