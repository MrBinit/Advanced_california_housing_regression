```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting the numerical columns and 'SalePrice'
selected_data = numerical_data[['LotArea', 'OverallQual', 'YearBuilt', 'TotalBsmtSF', 'GrLivArea', 'FullBath', 'GarageCars', 'SalePrice']]

# Plotting scatter plots for each numerical variable against 'SalePrice'
plt.figure(figsize=(16, 10))
for i, column in enumerate(selected_data.columns[:-1]):
    plt.subplot(3, 3, i+1)  # Creating subplots
    sns.scatterplot(x=column, y='SalePrice', data=selected_data)
    plt.title(f'Scatter plot of {column} vs SalePrice')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame named df
# Replace 'numerical_data' with your DataFrame and adjust 'feature_columns' if needed
feature_columns = ['LotArea', 'OverallQual', 'YearBuilt', 'TotalBsmtSF', 'GrLivArea', 'FullBath', 'GarageCars', 'SalePrice']
num_features = len(feature_columns)

# Calculate number of rows and columns for subplots
num_rows = (num_features + 1) // 2
num_cols = 2

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 12))
axes = axes.flatten()

# Loop through each feature and create density plot
for i, column in enumerate(feature_columns):
    ax = axes[i]
    sns.kdeplot(data=numerical_data[column], shade=True, ax=ax)
    ax.set_title(f'Density Plot of {column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Density')

# Remove any empty subplots
for i in range(num_features, num_rows*num_cols):
    fig.delaxes(axes[i])

# Adjust layout
plt.tight_layout()
plt.show()
```
